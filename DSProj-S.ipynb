```python
#Author: Soumya
#Task1 :facebook likes versus imdb score
#Analysis: It is a general assumption that movie having more facebook likes will have greater imdb score 
# but it is not the always the case. From the visualization illustrated below, we observe that movie having
# 1 lakh facebook likes has imdb score 9.5 whereas movies with 130000-200000 facebook likes have their imdb score 
#in range of 6.8 to 8.Thus facebook likes are not a good predictor of imdb score. Facebook likes are influenced by
#many factors such as popularity of actors or director of the movie, how well the movie was publicized on social media
#etc whereas the imdb score depends totally on the votes by registered imdb users.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%pylab inline

dataset=pd.read_csv('movie_metadata.csv')
df1=dataset[['movie_facebook_likes','imdb_score']]
df1 = df1[df1.movie_facebook_likes>0]
s = sns.lmplot('movie_facebook_likes', 'imdb_score',data=df1,fit_reg=False,size=10)
s.set(xlim=(0, None))
s.set(ylim=(0, None))
x_values = [10000,30000, 50000, 70000, 90000, 110000, 130000, 150000, 170000,200000,300000]
plt.xticks(x_values)
plt.title('Facebook likes versus IMDb score')
plt.ylabel('IMDB score')
plt.xlabel('Facebook Likes')
plt.show()


```

```python
#Author: Soumya
#Task2 : Content Rating versus imdb score
#Analysis: The movies involving voilence, sexual situations and crude language specifically designed for adults
# have highest imdb scores as compared to movies involving fantasy and animations designed for children. eg, MEAN 
# imdb scores for TV-PG(suitable for adults only) is 8.2. 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%pylab inline

dataset=pd.read_csv('movie_metadata.csv')
df1=dataset[['content_rating','imdb_score']]
df2=df1.groupby(by=['content_rating'])['imdb_score'].mean()
df2.plot(kind='bar',position=1,title='Content Rating versus Mean Imdb score',figsize=(12,7))



```
